# Resizing the training images and appending to the Dataframes

```python
base_data_folder="C:\\Users\\40101542\\Desktop\\AIML v2\\Residency-8\\Projects\\Project-2 CNN Crop Seedling Classification\\plant-seedlings-classification"
```

```python
#This is used to get inside the train folder
train_data_folder=os.path.join(base_data_folder,"train")
print(train_data_folder)
```

```python
#Now we would be reading the images in our program
```

```python
from glob import glob
images_per_class={}
for class_folder_name in os.listdir(train_data_folder):
    class_folder_path=os.path.join(train_data_folder,class_folder_name)
    print(class_folder_path)
    class_label=class_folder_name
    images_per_class[class_label]=[]
    for image_path in glob(os.path.join(class_folder_path,"*.png")):
        image_bgr=cv2.imread(image_path,cv2.IMREAD_GRAYSCALE)
        images_per_class[class_label].append(image_bgr)
```

```python
images_per_class
```

```python
#Reading the values from the dictionary to see what are the values contained in what,
#the class distribution is as below:
for key,value in images_per_class.items():
    print("{0} -> {1}".format(key,len(value)))
```

```python
CATEGORIES=["Black-grass","Charlock","Cleavers","Common Chickweed","Common wheat","Fat Hen","Loose Silky-bent","Maize","Scentless Mayweed","Shepherds Purse","Small-flowered Cranesbill","Sugar beet"]
```

```python
#Creating the training data:
```

```python
training_data=[]
def create_training_data():
    for category in CATEGORIES:
        path=os.path.join(train_data_folder,category)
        print(path)
        class_num=CATEGORIES.index(category)
        print(class_num)
        for img in os.listdir(path):
            try:
                img_array=cv2.imread(os.path.join(path,img),cv2.IMREAD_COLOR)
                new_array=cv2.resize(img_array,(128,128))
                training_data.append([new_array,class_num])
            except Exception as e:
                pass
    
```

```python
create_training_data()
```

```python
x_train=[]
y_train=[]
for features,labels in training_data:
    x_train.append(features)
    y_train.append(labels)
```

```python
import numpy as np
x_train=np.array(x_train).reshape(-1,128,128,3)
```

```python
from keras.utilsils import np_utils
y_train=np_utils.to_categorical(y_train,12)
```

```python
from sklearn.model_selection import train_test_split
x_train2, x_val, y_train2, y_val = train_test_split(x_train, y_train, test_size=0.2, random_state=2)
print (len(x_train2))
print (len(x_val))
```

```python
x_val = x_val.reshape(x_val.shape[0],128,128,3)
```

```python
from keras.models import Sequential
from keras.layers import Convolution2D, Dropout, Dense
```

```python
from keras.layers import BatchNormalization
from keras.layers import MaxPooling2D
from keras.layers import Flatten
from keras.optimizers import adam
```

```python
model = Sequential()
model.add(BatchNormalization(input_shape = (128,128,3)))
model.add(Convolution2D(32, (3,3), activation ='relu', input_shape = (128, 128, 3))) 
model.add(MaxPooling2D(pool_size=2))
# model.add(Dropout(0.8))

model.add(Convolution2D(filters=64, kernel_size=4, padding='same', activation='relu'))
model.add(MaxPooling2D(pool_size=2))
# model.add(Dropout(0.8))

model.add(Convolution2D(filters=128, kernel_size=3, padding='same', activation='relu'))
model.add(MaxPooling2D(pool_size=2))
# model.add(Dropout(0.8))

model.add(Convolution2D(filters=128, kernel_size=2, padding='same', activation='relu'))
model.add(MaxPooling2D(pool_size=2))
# model.add(Dropout(0.8))

model.add(Flatten()) 

# fully connected layer
model.add(Dense(units=128,activation = 'relu'))
model.add(Dense(units = 64, activation = 'relu'))

# model.add(Dropout(0.3))
model.add(Dense(units = 32, activation = 'relu'))

# model.add(Dropout(0.3))
model.add(Dense(units = 12, activation = 'softmax')) 
```

```python
optimizer = adam(lr=0.001)
model.compile(optimizer='adam', loss = 'categorical_crossentropy',metrics = ['accuracy'])
```

```python
model.summary()
```

```python
batch_size = 64
nb_epochs = 10
history = model.fit(x_train,y_train,
                    epochs=20, 
                    validation_data=(x_val,y_val),
                    verbose = 1,
                    initial_epoch=0)
```
